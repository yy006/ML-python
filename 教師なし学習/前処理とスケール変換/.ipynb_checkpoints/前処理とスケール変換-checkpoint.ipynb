```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
 random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
print(cancer.keys()) 
print(cancer['target'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

- fitãƒ¡ã‚½ãƒƒãƒ‰ã‚’è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã«å¯¾ã—ã¦é©ç”¨ã—ã¦ã€ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›å™¨ã‚’é©åˆã•ã›ã‚‹  
- MinMaxScalerã®å ´åˆã¯ã€fitãƒ¡ã‚½ãƒƒãƒ‰ã¯è¨“ç·´ãƒ‡ãƒ¼ã‚¿ä¸­ã®å„ç‰¹å¾´é‡ã®æœ€å°å€¤ã¨æœ€å¤§å€¤ã‚’è¨ˆç®—ã™ã‚‹

```python
scaler.fit(X_train)
```

- å­¦ç¿’ã—ãŸå¤‰æ›ã‚’é©ç”¨ã™ã‚‹ã«ã¯ã€ã¤ã¾ã‚Šå®Ÿéš›ã«è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã‚’ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã™ã‚‹ã«ã¯ã€ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›å™¨ã®transformãƒ¡ã‚½ãƒƒãƒ‰ã‚’ç”¨ã„ã‚‹

```python
# ãƒ‡ãƒ¼ã‚¿ã‚’å¤‰æ›
X_train_scaled = scaler.transform(X_train)
# ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã®å‰å¾Œã®ãƒ‡ãƒ¼ã‚¿ç‰¹æ€§ã‚’ãƒ—ãƒªãƒ³ãƒˆ
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
X_train_scaled.max(axis=0)))
```

```python
# ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã‚’å¤‰æ›
X_test_scaled = scaler.transform(X_test)
# ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã®å‰å¾Œã®ãƒ‡ãƒ¼ã‚¿ç‰¹æ€§ã‚’ãƒ—ãƒªãƒ³ãƒˆ
print("per-feature minimum after scaling:\n{}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n{}".format(X_test_scaled.max(axis=0)))
```

- ãƒ†ã‚¹ãƒˆã‚»ãƒƒãƒˆã®å ´åˆã«ã¯ã€ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›å¾Œã®æœ€å°å€¤ã¨æœ€å¤§å€¤ãŒ0ã¨1ã«ãªã£ã¦ã„ãªã„ã€‚  
- MinMaxScalerãŒï¼ˆä»–ã®ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›å™¨ã‚‚åŒã˜ï¼‰å¸¸ã«è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã«ã¾ã£ãŸãåŒã˜å¤‰æ›ã‚’æ–½ã™  
- transformãƒ¡ã‚½ãƒƒãƒ‰ã¯å¸¸ã«è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã®æœ€å°å€¤ã‚’å¼•ãã€è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã®ãƒ¬ãƒ³ã‚¸ã§å‰²ã‚‹ã€‚ã“ã‚Œã‚‰ã®å€¤ã¯,ãƒ†ã‚¹ãƒˆã‚»ãƒƒãƒˆã®æœ€å°å€¤ã‚„ãƒ¬ãƒ³ã‚¸ã¨ã¯é•ã†å ´åˆã‚‚ã‚ã‚‹ã€‚

```python
from sklearn.datasets import make_blobs
# åˆæˆãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
# è¨“ç·´ã‚»ãƒƒãƒˆã¨ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã«åˆ†å‰²
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)
# è¨“ç·´ã‚»ãƒƒãƒˆã¨ãƒ†ã‚¹ãƒˆã‚»ãƒƒãƒˆã‚’ãƒ—ãƒ­ãƒƒãƒˆ
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1],
                 color=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^',
                 color=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")
# MinMaxScalerã§ãƒ‡ãƒ¼ã‚¿ã‚’ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
# ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã•ã‚ŒãŸãƒ‡ãƒ¼ã‚¿ã®ç‰¹æ€§ã‚’å¯è¦–åŒ–
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                 color=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^',
                 color=mglearn.cm2(1), label="Test set", s=60)
axes[1].set_title("Scaled Data")
# ãƒ†ã‚¹ãƒˆã‚»ãƒƒãƒˆã‚’è¨“ç·´ã‚»ãƒƒãƒˆã¨ã¯åˆ¥ã«ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›
# æœ€å°å€¤ã¨æœ€å¤§å€¤ãŒ0,1ã«ãªã‚‹ã€‚ã“ã“ã§ã¯èª¬æ˜Žã®ãŸã‚ã«ã‚ã–ã¨ã‚„ã£ã¦ã„ã‚‹
# ï¼Šå®Ÿéš›ã«ã¯ã‚„ã£ã¦ã¯ã„ã‘ãªã„ï¼ï¼Š
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)
# é–“é•ã£ã¦ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã•ã‚ŒãŸãƒ‡ãƒ¼ã‚¿ã‚’å¯è¦–åŒ–
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                 color=mglearn.cm2(0), label="training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1],
                 marker='^', color=mglearn.cm2(1), label="test set", s=60)
axes[2].set_title("Improperly Scaled Data")
for ax in axes:
     ax.set_xlabel("Feature 0")
     ax.set_ylabel("Feature 1")
```
